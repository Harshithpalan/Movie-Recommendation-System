Create a simple movie database system using pandas. This system should allow for initializing an empty database with columns like 'Title', 'Genre', 'Year', 'Director', and 'Rating'. ## Initialize Movie Database

### Subtask:
Create an empty pandas DataFrame to serve as the movie database, defining columns such as 'Title', 'Genre', 'Year', 'Director', and 'Rating'.


**Reasoning**:
First, I'll import the pandas library, which is essential for working with DataFrames in Python. Then, I'll create an empty pandas DataFrame with the specified columns ('Title', 'Genre', 'Year', 'Director', 'Rating') and assign it to the `movie_database` variable.



```python
import pandas as pd

columns = ['Title', 'Genre', 'Year', 'Director', 'Rating']
movie_database = pd.DataFrame(columns=columns)

print(movie_database)
```

## Add Movies to Database

### Subtask:
Implement a function to add new movie entries to the DataFrame. This function will take movie details as input and append them as a new row.


**Reasoning**:
I will define the `add_movie` function as specified, which will create a new DataFrame row from input parameters and append it to the `movie_database` using `pd.concat`. After defining the function, I will add two example movies and then print the updated `movie_database` to demonstrate its functionality.



```python
def add_movie(title, genre, year, director, rating):
    global movie_database
    new_movie = pd.DataFrame([{'Title': title, 'Genre': genre, 'Year': year, 'Director': director, 'Rating': rating}])
    movie_database = pd.concat([movie_database, new_movie], ignore_index=True)
    print(f"Added movie: {title}")

# Add example movies
add_movie('Inception', 'Sci-Fi', 2010, 'Christopher Nolan', 8.8)
add_movie('The Matrix', 'Sci-Fi', 1999, 'Lana Wachowski', 8.7)
add_movie('Interstellar', 'Sci-Fi', 2014, 'Christopher Nolan', 8.6)

# Display the updated database
print("\nUpdated Movie Database:")
print(movie_database)
```

**Reasoning**:
The `FutureWarning` arises from `pd.concat` when combining an empty DataFrame with a new one, as pandas might change its dtype inference behavior. To address this, I will modify the `add_movie` function to check if the `movie_database` is empty. If it is, the new movie DataFrame will directly initialize `movie_database` to establish correct dtypes from the start, avoiding the warning on subsequent concatenations.



```python
def add_movie(title, genre, year, director, rating):
    global movie_database
    new_movie_data = {'Title': title, 'Genre': genre, 'Year': year, 'Director': director, 'Rating': rating}
    new_movie_df = pd.DataFrame([new_movie_data])

    if movie_database.empty:
        movie_database = new_movie_df
    else:
        movie_database = pd.concat([movie_database, new_movie_df], ignore_index=True)

    print(f"Added movie: {title}")

# Add example movies
add_movie('Inception', 'Sci-Fi', 2010, 'Christopher Nolan', 8.8)
add_movie('The Matrix', 'Sci-Fi', 1999, 'Lana Wachowski', 8.7)
add_movie('Interstellar', 'Sci-Fi', 2014, 'Christopher Nolan', 8.6)

# Display the updated database
print("\nUpdated Movie Database:")
print(movie_database)
```

**Reasoning**:
The previous execution of the code block resulted in duplicate entries because the `movie_database` was not reset before re-adding the example movies. To correct this and ensure the database reflects only the movies added by the current function, I will first re-initialize the `movie_database` to an empty DataFrame and then re-add the movies using the corrected `add_movie` function.



```python
import pandas as pd

# Re-initialize the movie_database to clear previous entries and ensure correct state
columns = ['Title', 'Genre', 'Year', 'Director', 'Rating']
movie_database = pd.DataFrame(columns=columns)

def add_movie(title, genre, year, director, rating):
    global movie_database
    new_movie_data = {'Title': title, 'Genre': genre, 'Year': year, 'Director': director, 'Rating': rating}
    new_movie_df = pd.DataFrame([new_movie_data])

    if movie_database.empty:
        movie_database = new_movie_df
    else:
        movie_database = pd.concat([movie_database, new_movie_df], ignore_index=True)

    print(f"Added movie: {title}")

# Add example movies
add_movie('Inception', 'Sci-Fi', 2010, 'Christopher Nolan', 8.8)
add_movie('The Matrix', 'Sci-Fi', 1999, 'Lana Wachowski', 8.7)
add_movie('Interstellar', 'Sci-Fi', 2014, 'Christopher Nolan', 8.6)

# Display the updated database
print("\nUpdated Movie Database:")
print(movie_database)
```

## Search and Filter Movies

### Subtask:
Implement functions to search for movies by title or filter them by genre, year, or rating, demonstrating how to query the database.


**Reasoning**:
I will implement the `search_movie_by_title` function for case-insensitive title search and the `filter_movies` function for filtering by genre, year, or rating, then demonstrate their usage with example searches and filters.



```python
def search_movie_by_title(title):
    if movie_database.empty:
        print("Movie database is empty.")
        return pd.DataFrame(columns=movie_database.columns)
    # Case-insensitive search for title
    mask = movie_database['Title'].str.contains(title, case=False, na=False)
    return movie_database[mask]

def filter_movies(genre=None, year=None, rating_threshold=None):
    if movie_database.empty:
        print("Movie database is empty.")
        return pd.DataFrame(columns=movie_database.columns)

    filtered_df = movie_database.copy()

    if genre:
        filtered_df = filtered_df[filtered_df['Genre'].str.contains(genre, case=False, na=False)]
    if year is not None:
        filtered_df = filtered_df[filtered_df['Year'] == year]
    if rating_threshold is not None:
        filtered_df = filtered_df[filtered_df['Rating'] >= rating_threshold]

    return filtered_df

# 3. Demonstrate search_movie_by_title function
print("\n--- Searching for 'Inception' ---")
search_results = search_movie_by_title('Inception')
if not search_results.empty:
    print(search_results)
else:
    print("No movies found matching 'Inception'.")

print("\n--- Searching for 'Matrix' ---")
search_results = search_movie_by_title('Matrix')
if not search_results.empty:
    print(search_results)
else:
    print("No movies found matching 'Matrix'.")

# 4. Demonstrate filter_movies by genre
print("\n--- Filtering by Genre: 'Sci-Fi' ---")
genre_filter_results = filter_movies(genre='Sci-Fi')
if not genre_filter_results.empty:
    print(genre_filter_results)
else:
    print("No movies found for 'Sci-Fi' genre.")

# 5. Demonstrate filter_movies by year
print("\n--- Filtering by Year: 2010 ---")
year_filter_results = filter_movies(year=2010)
if not year_filter_results.empty:
    print(year_filter_results)
else:
    print("No movies found from year 2010.")

# 6. Demonstrate filter_movies by rating threshold
print("\n--- Filtering by Rating >= 8.7 ---")
rating_filter_results = filter_movies(rating_threshold=8.7)
if not rating_filter_results.empty:
    print(rating_filter_results)
else:
    print("No movies found with rating >= 8.7.")

print("\n--- Filtering by Genre='Sci-Fi' AND Year=2014 AND Rating>=8.5 ---")
combined_filter_results = filter_movies(genre='Sci-Fi', year=2014, rating_threshold=8.5)
if not combined_filter_results.empty:
    print(combined_filter_results)
else:
    print("No movies found matching all criteria.")
```

## Display Movie Information

### Subtask:
Create a function to display all movies or the results of a search/filter operation in a readable format.


**Reasoning**:
I will define the `display_movies` function, as specified, to handle both empty and non-empty DataFrames, and then demonstrate its usage by displaying the full `movie_database` and the `search_results` from the previous step.



```python
def display_movies(movies_df, message="--- Movie Listings ---"):
    if movies_df.empty:
        print("No movies to display.")
    else:
        print(message)
        print(movies_df)

# Display the entire movie database
print("\n--- Displaying entire movie database ---")
display_movies(movie_database)

# Display a specific search result (e.g., from 'search_results' for 'Inception')
# Re-running the search to ensure search_results is current
search_results_inception = search_movie_by_title('Inception')
print("\n--- Displaying search results for 'Inception' ---")
display_movies(search_results_inception, message="--- Search Results for 'Inception' ---")

# Display a specific filter result (e.g., from 'genre_filter_results' for 'Sci-Fi')
# Re-running the filter to ensure genre_filter_results is current
genre_filter_results_scifi = filter_movies(genre='Sci-Fi')
print("\n--- Displaying filter results for 'Sci-Fi' genre ---")
display_movies(genre_filter_results_scifi, message="--- Filter Results for 'Sci-Fi' Genre ---")
```
